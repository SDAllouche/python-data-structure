```python
class Node :
    
    def __init__(self,data=None,Next=None):
        self.data=data
        self.next=Next
```

```python
class LinkedList:
    
    def __init__(self):
        self.head=None
    
    def addFirst (self,data):
        node=Node(data,self.head)
        self.head=node
        
    def addEnd(self,data):
        if self.head is None:
            self.head=Node(data,None)
            return
        else:
            itr=self.head
            while itr.next:
                itr=itr.next
            itr.next=Node(data,None)
            
    def size(self):
        count = 0
        itr = self.head
        while itr:
            count+=1
            itr = itr.next
        return count
            
    def addAt(self,index,data):
        if index<0  :
            return "Invalid Index"
        if index>self.size():
            self.addEnd(data)
        if self.head is None or index==0:
            self.addFirst (data)
        else:
            itr=self.head
            i=0
            while itr:
                if i==index-1:
                    node=Node(data,itr.next)
                    itr.next=node
                    return
                itr = itr.next
                i=i+1
                
    def removeFirst(self,index):
        if index<0 or index>=self.size():
            return "Invalid Index"

        if index==0:
            self.head = self.head.next
            return

        count = 0
        itr = self.head
        while itr:
            if count == index - 1:
                itr.next = itr.next.next
                return
            itr = itr.next
            count+=1
            
    def removeEnd(self,index):
        if index<0 or index>=self.size():
            return "Invalid Index"

        if index==0:
            self.head = self.head.next
            return

        count = 0
        itr = self.head
        while itr:
            if count == index - 1:
                itr.next = itr.next.next
                return
            itr = itr.next
            count+=1
                
    def removeAt(self,index):
        if index<0 or index>=self.size():
            return 'Invalid Index'
        if index==0:
            self.head = self.head.next
            return
        count = 0
        itr = self.head
        while itr:
            if count == index - 1:
                itr.next = itr.next.next
                return
            itr = itr.next
            count+=1
            
    def research(self,x):
        if self.head is None:
            return 'Liste is empty'
        else:
            itr=self.head
            position=0
            while itr:
                if itr.data==x:
                    return "Position : " + str(position+1)
                itr=itr.next
                position=position+1
                    
            return 'Not found'
    
    def toString(self):
        if self.head is None:
            return 'Liste is empty'
        else:
            itr=self.head
            elem=''
            while itr:
                elem+=str(itr.data)+" --> " if itr.next!=None else str(itr.data)
                itr=itr.next
            print(elem)
```
