```python
#Data
students= ['Sara', 'Zineb', 'Mohamed', 'Ali', 'Khadija', 'Idriss', 'Najat', 'Nadia', 'Marouane', 'Ahmed']
#---------------------------------------------------------------------------------------------------
genders  = [   'F'   ,   'F' ,  'M'  ,   'M' ,  'F'  ,    'M'  ,  'F'   ,   'F'  ,  'M'  ,  'M' ]
groups  = [    3    ,    1  ,   1   ,    3  ,   3   ,     1   ,   2    ,    2   ,   1   ,   3 ]
regions=['S','C','N']

#---------------------------------------------------------------------------------------------------
#scores   = {'DS': length of DS is the number of DS
#                  The element of each DS is the mark of each student
#           'TP': length of TP is the number of TP
#                  The element of each TP is the mark of each student
scores   = {'DS' : [[36, 58, 46, 96, 9, 82, 83, 66, 35, 47],
                   [46, 50, 55, 21, 22, 76, 51, 90, 96, 48],
                   [56, 54, 53, 17, 31, 74, 11, 53, 98, 67],
                   [77, 38, 8, 74, 39, 39, 52, 66, 38, 86],
                   [93, 21, 7, 33, 10, 97, 48, 96, 24, 7],
                   [97, 98, 95, 75, 64, 9, 48, 51, 45, 82]],

           'TP': [[48, 63, 98, 47, 25, 90, 100, 21, 41, 44],
                  [73, 79, 78, 39, 11, 100, 57, 96, 13, 99]]}
#---------------------------------------------------------------------------------------------------
# attendance = { Day : [ Duration_session : [ delay of 10 sessions]]}
attendance = {'Monday'    : [60, [3, 31, 1, 56, 15, 49, 48, 20, 2, 38]],
              'Tuesday'   : [120,[8, 3, 2, 7, 5, 9, 4, 0, 6, 1]],
              'Wednesday' : [60, [2, 7, 14, 6, 9, 15, 12, 13, 10, 0]],
              'Friday'    : [120,[12, 50, 0, 15, 11, 10, 3, 6, 14, 2]]}
#---------------------------------------------------------------------------------------------------
# min_att = {Attendance_statut : Late_Minute}
min_att = {'Present': range(0,1,1), 'Late' : range(1,6,1), 'Very Late': range(6,15,1), 'Absent': range(15,180,1)} 
# dic_att = { Attendance_statut : rate}
dic_att = {'Present': 1.0, 'Late': 0.97, 'Very Late': 0.93, 'Absent': 0.9} 
dic_all_data={'students':students,'scores':scores,'attendance':attendance}
```

```python
#%%
# Q1. Print the name of members of the group 3.

def names_By_Group(group_number):
    return [students[i] for i in range(len(students)) if groups[i]==group_number]

print("Q1 : \n",names_By_Group(3))
```

```python
#%%
# Q2. Print the number of each gender in each group.

def gender_in_group():
    gender={i+1:{'F':0,'M':0} for i in range(len(set(groups)))}
    for i,j in zip(genders,groups):
        gender[j][i]+=1
    return gender

print("Q2 : \n",gender_in_group())
```

```python
#%%
# Q3. Print the name of students who were absent for Monday seance.

def names_by_absent(day,type_attendence):
    return [students[i] for i in range(len(students)) if attendance[day][1][i] in min_att[type_attendence]]

print("Q3 : \n",names_by_absent('Monday','Absent'))
```

```python
#%%
# Q4. Print the name of students with more than 50 points in all DS.

def names_by_points(score):
    print("Q4 :")
    for i in range(len(students)):
        if scores['DS'][0][i]>score and scores['DS'][1][i]>score and scores['DS'][2][i]>score and scores['DS'][3][i]>score and scores['DS'][4][i]>score and scores['DS'][5][i]>score:
            print(students[i])
            
names_by_points(50)
```

```python
#%%
# Q5. Print the mean mark (DS) of all student of each group.

def mean_by_group():
    dictionary={i+1:0 for i in range(len(set(groups)))}
    for i in range(len(students)):
        student_mean=(scores['DS'][0][i]+scores['DS'][1][i]+scores['DS'][2][i]+scores['DS'][3][i]+scores['DS'][4][i]+scores['DS'][5][i])/6
        group_mean=student_mean if dictionary[groups[i]]==0 else (student_mean+dictionary[groups[i]])/2
        dictionary[groups[i]]=round(group_mean,2)
    return dictionary
    
print("Q5 : \n",mean_by_group())
```

```python
#%%
# Q6. Print the seance for each group with less delay in min.

def seance_with_less_delay():
    attendance_by_group={1:{},2:{},3:{}}
    print("Q6 :")
    for key,value in attendance.items():
        g1=0;g2=0;g3=0
        for i in range(len(students)):
            if groups[i]==1:
                g1+=value[1][i]
            elif groups[i]==2:
                g2+=value[1][i]
            else :
                g3+=value[1][i]
                
        attendance_by_group[1][key]=g1
        attendance_by_group[2][key]=g2
        attendance_by_group[3][key]=g3
        
    print("Seance with least delay for group 1 is :",min(attendance_by_group[1],key=attendance_by_group[1].get))
    print("Seance with least delay for group 2 is :",min(attendance_by_group[2],key=attendance_by_group[2].get))
    print("Seance with least delay for group 3 is :",min(attendance_by_group[3],key=attendance_by_group[3].get))

seance_with_less_delay()
```

```python
#%%
# Q7. Print the name of the student who has the lower mark (of all the DS) in each gender. 

def low_mark_by_gender():
    dictionary={'F':('',0),'M':('',0)}
    for i,name in enumerate(students):
        student_score=scores['DS'][0][i]+scores['DS'][1][i]+scores['DS'][2][i]+scores['DS'][3][i]+scores['DS'][4][i]+scores['DS'][5][i]
        if genders[i]=='F':
            if dictionary['F'][1]>student_score or dictionary['F'][1]==0:
                dictionary['F']=(name,student_score) 
        elif genders[i]=='M':
            if dictionary['M'][1]>student_score or dictionary['M'][1]==0:
                dictionary['M']=(name,student_score) 
    
    return dictionary

print("Q7 : \n",low_mark_by_gender())
```

```python
#%%
# Q8. Print the mean mark of TPs for each gender.

def mean_by_gender():
    dictionary={'F':0,'M':0}
    f,m=(0,0)
    for i in range(len(students)):
        student_mean=(scores['TP'][0][i]+scores['TP'][1][i])/2
        f,m=(f+1,m) if genders[i]=='F' else (f,m+1)
        dictionary[genders[i]]+=student_mean
    
    dictionary['F']=dictionary['F']/f
    dictionary['M']=dictionary['M']/m
    
    return dictionary

print("Q8 : \n",mean_by_gender())
```

```python
#%%
# Q9. Print the name of the student who has the lower minutes of delay in all seances.

attend=list(attendance.values())
days=len(attend)
marques=[(sum(attend[j][1][i] for j in range(days)),name) for i,name in enumerate(students)]
print("Q9 : \n",min(marques))
```

```python
#%%
# Q10. Print the name of  the student with a mean mark more than 80 in all TPs and an Attendance Rate more than 0.95.

def name_by_mark_rate(mark,rate):
    dictionary={}
    attend=list(attendance.values())
    for i,name in enumerate(students):
        m=(scores['TP'][0][i]+scores['TP'][1][i])/2
        student_attendance=attend[0][1][i]+attend[1][1][i]+attend[2][1][i]+attend[3][1][i]
        for key,value in min_att.items():
            r=dic_att[key] if student_attendance in value else 0
        if m>mark and r>rate :
            dictionary[name]=(m,r)
            
    return dictionary
    
print("Q10 : \n",name_by_mark_rate(80,0.95))
```

```python
#%%
# Q11. Print the name of the student who has the higher minutes of delay in the third seance.

def name_by_higher_delay(seance):
    
    print("Q11 :")
    index=attendance[seance][1].index(max(attendance[seance][1]))
    print(students[index])
    
name_by_higher_delay('Wednesday')  
```

```python
#%%
# Q12. Print the name of the Male student in group 1 with the highest note in the second DS.

def name_by_notes(gender,group,ds):
    
    print("Q12 :")
    maxim=('',0)
    for i,name in enumerate(students):
        maxim=(name,scores['DS'][ds-1][i]) if maxim[1]<scores['DS'][ds-1][i] and genders[i]==gender and groups[i]==group else maxim
    print(maxim)
    
name_by_notes('M',1,2)
```

```python
#%%
# Q13. Print the name of the students who progress after each DS.

def progress(index):
    prog=0
    for i in range(1,len(scores['DS'])):
        if scores['DS'][i][index]>scores['DS'][i-1][index]:
            prog+=1
    return prog
            
print("Q13 :")
for i,name in enumerate(students):
    if progress(i)==len(scores['DS'])-1:
        print(name)
```

```python
#%%
# Q14. Print the name of the student with the highest number of attendance for both Monday & Friday seances.

def name_by_attendance(*args):
    print("Q14 :")
    students_attendance={name:[attendance[j][0]-attendance[j][1][i] for j in args] for i,name in enumerate(students)}  
    print(max(students_attendance,key=students_attendance.get))
    
name_by_attendance('Monday','Friday')
```

```python
#%%
# Q15. Print the Finale_note for each student.

#  * Finale_note = (Mean_of_DS *.6 + Mean_of_TPs * 0.15 + Median_of_group_DS * 0.25) * Attendance_rate

def mean_of_DS(index) :
    mean=0
    for i in scores['DS']:
        mean+=i[index]
    return mean/len(scores['DS'])

def mean_of_TP(index):
    mean=0
    for i in scores['TP']:
        mean+=i[index]
    return mean/len(scores['TP'])

def Median_of_group_DS(group):
    liste=[mean_of_DS(i) for i in range(len(students)) if groups[i]==group]
    liste.sort()
    return ((liste[len(liste)//2])-1+liste[len(liste)//2])/2 if len(liste)%2==0 else liste[len(liste)//2]

def attendance_rate(index):
    student_attendance=0
    for i in attendance.values():
        student_attendance+=i[1][index]
    for key,value in min_att.items():
        rate=dic_att[key] if student_attendance in value else 0   
    return rate

print("Q15 : ")
for i,name in enumerate(students):
    note=(mean_of_DS(i)*0.6 + mean_of_TP(i)*0.15 + Median_of_group_DS(groups[i])*0.25 )*attendance_rate(i)
    print("Final note of " + name + " is : " + str(round(note,3)))
```
