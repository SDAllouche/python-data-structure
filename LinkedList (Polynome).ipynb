```python
class Monome :
    
    def __init__(self,exp=None,coef=None,Next=None):
        self.exp=exp
        self.coef=coef
        self.next=Next
```

```python
class Polynome:
    
    def __init__(self):
        self.head=None
        
    def create(self,exp,coef):
        if exp>=0:
            monome=Monome(exp,coef,self.head)
            self.head=monome
            
    def createReverse(self,exp,coef):
        if self.head is None:
            self.head=Monome(exp,coef,None)
            return
        else:
            itr=self.head
            while itr.next:
                itr=itr.next
            itr.next=Monome(exp,coef,None)
            
    def affiche(self):
        if self.head is None:
            return 'Empty'
        else:
            itr=self.head
            elem=''
            while itr:
                if itr.coef==0:
                    itr=itr.next
                    continue
                elif itr.coef==1:
                    if itr.exp==0:
                        elem+="1"+" + " if itr.next!=None else "1"
                    elif itr.exp==1:
                        elem+="x"+" + " if itr.next!=None else "x"
                    elif itr.exp>1:
                        elem+="x**"+str(itr.exp)+" + " if itr.next!=None else "x**"+str(itr.exp)
                elif itr.coef==-1:
                    if itr.exp==0:
                        elem+="-1"+" + " if itr.next!=None else "-1"
                    elif itr.exp==1:
                        elem+="-x"+" + " if itr.next!=None else "x"
                    elif itr.exp>1:
                        elem+="-x**"+str(itr.exp)+" + " if itr.next!=None else "-x**"+str(itr.exp)
                else:
                    if itr.exp==0:
                        elem+=str(itr.coef)+" + " if itr.next!=None else str(itr.coef)
                    elif itr.exp==1:
                        elem+=str(itr.coef)+"x"+" + " if itr.next!=None else str(itr.coef)+"x"
                    elif itr.exp>1:
                        elem+=str(itr.coef)+"x**"+str(itr.exp)+" + " if itr.next!=None else str(itr.coef)+"x**"+str(itr.exp)
                itr=itr.next
            if elem.endswith("+ "):
                elem=elem[:-2]
            print(elem.replace("+ -","-  "))
            
    def addition(self,P,Q):
        p=P.head
        q=Q.head
        if p.exp>q.exp:
            itr=p
        if p.exp<=q.exp:
            itr=q
        while itr:
            if p.exp==q.exp:
                self.createReverse(p.exp,p.coef+q.coef)
                itr=itr.next
                p=p.next
                q=q.next
            elif p.exp>q.exp:
                self.createReverse(p.exp,p.coef)
                itr=itr.next
                p=p.next
            elif p.exp<q.exp:
                self.createReverse(q.exp,q.coef)
                itr=itr.next
                q=q.next
                
    def multiplication(self,P,Q):
        tempPoly=Polynome()
        p=P.head
        while p:
            q=Q.head
            while q:
                tempPoly.createReverse(p.exp+q.exp,p.coef*q.coef)
                q=q.next
            p=p.next
        
        for i in range(tempPoly.head.exp+1):
            itr=tempPoly.head
            newCoef=0
            while itr:
                if i==itr.exp:
                    newCoef+=itr.coef
                itr=itr.next
            self.create(i,newCoef)
```

```python
liste=Polynome()
liste.create(0,-1)
liste.create(1,2)
liste.create(2,3)
liste.create(3,-4)
liste.affiche()
```

```python
qliste=Polynome()
qliste.create(0,1)
qliste.create(1,2)
qliste.create(2,3)
qliste.create(3,4)
qliste.create(4,1)
qliste.affiche()
```

```python
hliste=Polynome()
hliste.addition(liste,qliste)
hliste.affiche()
```

```python
mliste=Polynome()
mliste.multiplication(liste,qliste)
mliste.affiche()
```
